```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dummy-full.csv', encoding = 'utf8', sep=',', error_bad_lines=False)
```

```python
df.head()
```

```python
df['pm10'] = pd.to_numeric(df['pm10'])
df['pm2_5'] = pd.to_numeric(df['pm2_5'])
```

```python
df_pivot_25 = df.pivot_table(index = ['time'], columns=['id'], values = 'pm2_5', aggfunc=np.mean)
df_pivot_10 = df.pivot_table(index = ['time'], columns=['id'], values = 'pm10', aggfunc=np.mean)
```

```python
plt.matshow(df_pivot_10.corr())
```

```python
df_corr = df_pivot_10.corr()
```
